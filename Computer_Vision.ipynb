```python
import cv2
cv2.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
img = cv2.imread('stary_night.jpg')
plt.imshow(img)
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Convert the image into gray scale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray')
plt.show()
```

```python
# Transform the image into HSV and HLS models(Hue,saturation,value/ hue,lightness,value)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
```

```python
# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
# Copy the image
img_copy = img.copy()
# Draw a rectangle 
cv2.rectangle(img_copy, pt1 = (400, 470), pt2 = (580, 530), 
              color = (255, 0, 0), thickness = 5)
plt.imshow(img_copy)
```

```python
# Draw a circle 
cv2.circle(img_copy, center = (500, 50), radius = 50, 
           color = (0, 0, 255), thickness = 5)
plt.imshow(img_copy)
```

```python
# Add text 
cv2.putText(img_copy, text = "Stary Night", 
            org = (250, 250),
            fontFace = cv2.FONT_HERSHEY_DUPLEX, 
            fontScale = 2, 
            color = (0, 255, 0), 
            thickness = 2, 
            lineType = cv2.LINE_AA)
plt.imshow(img_copy)
```
